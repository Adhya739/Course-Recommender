# Popularity-Based Course Recommendation

This notebook recommends courses based on number of subscribers/reviews.

```python
import pandas as pd

# Load dataset
courses = pd.read_csv("../data/udemy_course_data.csv")

def popular_courses(top_n=5):
    return courses.sort_values("num_subscribers", ascending=False).head(top_n)[["course_title","num_subscribers"]]

popular_courses()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import os

# Load dataset
# The original path '../data/udemy_course_data.csv' resulted in a FileNotFoundError.
# Please ensure 'udemy_course_data.csv' is uploaded to your Colab session's
# root directory or specify the correct path where the file is located.
try:
    courses = pd.read_csv("udemy_course_data.csv")
    file_loaded = True
except FileNotFoundError:
    print("Error: 'udemy_course_data.csv' not found.")
    print("Please upload the 'udemy_course_data.csv' file to your Colab session's root directory (e.g., by dragging and dropping it into the files pane).")
    courses = pd.DataFrame(columns=['course_title']) # Create an empty DataFrame with the expected column to prevent further errors
    file_loaded = False

# Proceed with preprocessing and recommendations only if the file was loaded successfully and is not empty
if file_loaded and not courses.empty:
    # Preprocess
    courses["course_title"] = courses["course_title"].str.lower()

    # TF-IDF
    tfidf = TfidfVectorizer(stop_words="english")
    tfidf_matrix = tfidf.fit_transform(courses["course_title"])

    # Cosine similarity
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

    def recommend_course(idx, top_n=5):
        if idx < 0 or idx >= len(courses):
            print(f"Error: Course index {idx} is out of bounds for the loaded data (0 to {len(courses) - 1}).")
            return []
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        # Ensure we don't try to access elements beyond the length of sim_scores
        # sim_scores[0] is the course itself, so we take from index 1.
        end_idx = min(len(sim_scores), top_n + 1)
        return [courses.iloc[i[0]]["course_title"] for i in sim_scores[1:end_idx]]

    # Call the recommendation function and print results
    print(recommend_course(10))
else:
    print("Course recommendation system not initialized due to missing data file or empty dataset.")
```
